```python
from matplotlib import style
style.use("ggplot")
# plt.plot(x,y,z)
days = [1,2,3,4,5]
temp = [36,35,37.7,38,40.2]
plt.plot(days,temp,marker="*",linewidth=7)
plt.xlabel('days',fontsize=15)
plt.ylabel('temperature',fontsize=15)
plt.title("1D Graph Rizwan")
plt.show()
```

```python
vijtemp = [40.5,43.5,38.5,39.2,38]
plt.plot(days,vijtemp,"m*--",linewidth=4,label="vja")
plt.plot(days,temp,color="g",marker=0,linewidth=4,label="gtr")
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Temp Graph b/w vja and ajv")
plt.legend(loc=2)
plt.grid(color='brown',linestyle="-",linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors
 
# Prepare a list of integers
val = [2, 3, 6, 9, 14]
 
# Prepare a list of sizes that increases with values in val
sizevalues = [i**2*50+50 for i in val]
 
# Prepare a list of colors
plotcolor = ['red','orange','yellow','green','blue']
 
# Draw a scatter plot of val points with sizes in sizevalues and
# colors in plotcolor
plt.scatter(val, val, s=sizevalues, c=plotcolor)
 
# Draw grid lines with red color and dashed style
plt.grid(color='red', linestyle='-.', linewidth=0.7)
 
# Set axis limits to show the markers completely
plt.xlim(0, 20)
plt.ylim(0, 20)
 
plt.show()
```

### HISTOGRAM

- represents as bins
- bins represent data points in range
- plt.hist(ele,bins=count)

```python
import numpy as np
std1 = np.random.randint(10,45,100)
std2 = np.random.randint(15,40,100)
plt.hist(std1,bins=100)
plt.xlabel('age')
plt.ylabel('no of student')
plt.show()
```

```python
plt.hist(std2,bins=100,color="c")
```

```python
vals = [num for num in range(10,50,3)]
plt.hist(std1,vals,histtype='bar',orientation="vertical",color="b")
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(8,5))
plt.hist([std1,std2],vals,rwidth=0.7,histtype="bar",orientation="vertical",color=["m","y"])
plt.title("Multiple Histograms")
plt.show()
```

### PIE PLOT

```python
x=[20,10,35,22,12]
y = [10,20,30,40,50,60,70]
plt.pie(x)
plt.show()
plt.pie(y)
plt.show()
```

```python
ex=[0,1,0,0,1]
f = ["average","good","above average","excellent","outstanding"]
plt.pie(x,labels=f,colors=['green','orange','blue','pink',"c"],autopct='%.f%%', frame=True,shadow=True,startangle=180,labeldistance=1.1,counterclock=True,explode=ex)
plt.title("Sort by Responses")
plt.show()
```
